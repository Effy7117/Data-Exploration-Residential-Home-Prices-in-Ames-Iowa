### Removing the null values

```python
# Checking the number of null values of each column
df.isnull().sum()[df.isnull().sum() !=0]
```

```python
# LotFrontage: Linear feet of street connected to property. 
# Imputing null values with mean
LotFrontage_mean = df.LotFrontage.mean()
df.LotFrontage.fillna(LotFrontage_mean, inplace = True)

# Alley: Type of alley access
# Impute with custom value 'Unknown'
df['Alley'].fillna('Unknown', inplace=True)

# MasVnrType: Masonry veneer type
# Impute with custom value 'unknown'
df['MasVnrType'].fillna('Unknown', inplace=True)

# MasVnrArea: Masonry veneer area in square feet
# Imputing null values with mean
MasVnrArea_mean = df.MasVnrArea.mean()
df.MasVnrArea.fillna(MasVnrArea_mean, inplace = True)

# BsmtQual: Height of the basement
# Imputing with custom value 'unknown'
df['BsmtQual'].fillna('Unknown', inplace=True)

# BsmtCond: General condition of the basement
# Imputing with custom value 'unknown'
df['BsmtCond'].fillna('Unknown', inplace=True)

# BsmtExposure: Walkout or garden level basement walls
# Imputing with custom value 'unknown'
df['BsmtExposure'].fillna('Unknown', inplace=True)

# BsmtFinType1: Quality of basement finished area
# Imputing with custom value 'unknown'
df['BsmtFinType1'].fillna('Unknown', inplace=True)

# BsmtFinType2: Quality of second finished area (if present)
# Imputing with custom value 'unknown'
df['BsmtFinType2'].fillna('Unknown', inplace=True)

# Electrical: Electrical system
# Just one missing values, hence removing the null value
df = df[~df.Electrical.isnull()].copy()
df.Electrical.isnull().sum()

# FireplaceQu: Fireplace quality
# Imputing with custom value 'unknown'
df['FireplaceQu'].fillna('Unknown', inplace=True)

# GarageType: Garage location
# Imputing with custom value 'unknown'
df['GarageType'].fillna('Unknown', inplace=True)

# GarageYrBlt: Year garage was built
# Average of YearBuilt and YrSold
GarageYrBlt_mean = df.GarageYrBlt.mean()
df.GarageYrBlt.fillna(GarageYrBlt_mean, inplace = True)

# GarageFinish: Interior finish of the garage
# Imputing with custom value 'unknown'
df['GarageFinish'].fillna('Unknown', inplace=True)

# GarageQual: Garage quality
# Imputing with custom value 'unknown'
df['GarageQual'].fillna('Unknown', inplace=True)

# GarageCond: Garage condition
# Imputing with custom value 'unknown'
df['GarageCond'].fillna('Unknown', inplace=True)

# PoolQC: Pool quality
# Imputing with custom value 'unknown'
df['PoolQC'].fillna('Unknown', inplace=True)

# Fence: Fence quality
# Imputing with custom value 'unknown'
df['Fence'].fillna('Unknown', inplace=True)

# MiscFeature: Miscellaneous feature not covered in other categories
# Imputing with custom value 'none'
df['MiscFeature'].fillna('none', inplace=True)
```

```python
# Verifying the number of null values of each column
df.isnull().sum()[df.isnull().sum() !=0]
```

### Correlation between numerical features and sale price

```python
# Exclude cztegorical columns from the correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Extract the correlation of each feature with the target variable "SalePrice"
correlation_with_saleprice = correlation_matrix["SalePrice"].abs().sort_values(ascending=False)

# Get the top 8 columns with the highest correlation
top_8_columns = correlation_with_saleprice[1:9].index
print("Top 8 correlated columns with SalePrice:")
for col in top_8_columns:
    correlation_value = correlation_matrix.loc[col, "SalePrice"]
    print(f"{col}: {correlation_value}")

top_8_column = ['SalePrice'] + list(top_8_columns)
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df[top_8_column].corr(), cmap="RdBu", annot=True, fmt=".2f")
plt.title("Correlations Between Top 8 Numerical Variables and SalePrice", size=10)
plt.show()
```

**Findings:**

**Numeric-columns with highest correlation with SalePrice:**
- OverallQual(Overall material and finish quality) : Correlation = 0.79 
- GrLivArea(Above grade (ground) living area square feet) : Correlation = 0.7 
- GarageCars(Size of garage in car capacity) : Correlation = 0.64 
- GarageArea(Size of garage in square feet) : Correlation = 0.62 
- TotalBsmtSF(Total square feet of basement area) : Correlation = 0.61 
- 1stFlrSF(First Floor square feet) : Correlation = 0.6 
- FullBath(Full bathrooms above grade) : Correlation = 0.56 
- TotRmsAbvGrd(Total rooms above grade (does not include bathrooms)) : Correlation = 0.53 

### Correlation between categorical features and sale price

```python
# Select categorical columns
cat_cols = df.select_dtypes(include='object').columns.tolist()

# Add the target variable to the list
cat_cols.append("SalePrice")

# Create a new DataFrame with selected columns
df_selected = df[cat_cols]

# Encode categorical variables into numerical format
df_encoded = pd.get_dummies(df_selected, drop_first=True)

# Calculate the correlation matrix
correlation_matrix = df_encoded.corr()

# Extract the correlation of each feature with the target variable "SalePrice"
correlation_with_saleprice = correlation_matrix["SalePrice"].sort_values(ascending=False)
```

- As Total square feet of basement area and First Floor square feet increases,  sale price increases and then decreases.
